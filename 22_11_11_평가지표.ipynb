정확도 = 맞춘 것으로만 평가

정밀도 = P라고 **예측**한 것 중 맟춘 비율

재현율 = **실제** P인 것 중 맞춘 P

F1 스코어 = 정밀도와 재현율을 평균 내겠다

Precision -> 예측을 함부로 못해 

Recall -> 아무거나 다 맞다고 하면 점수가 높아짐

---

두 개 다 써야겠구나...

둘 잘했을 때를 원한다.

### 산술 평균 = $ \frac{precision + recall} {2} $: 편향을 못 잡음

### 조화 평균 = $ \frac{2 * precision * recall}{precision + recall} $ : 한쪽이 0점

X 축 : 맞다고 예측했는데 틀린 비율 (FP의 비율)

Y 축 : 맞다고 예측했는데 맞춘 비율 (TP의 비율)

면적이 큰게 좋다 -> Y축에 붙으면 좋다.

그런데 비율이 커진다는게 이해안된다.

---

TPR(%) : 비율, 분모 = 실제P 즉 Recall의 다른 말 (sesitivity)

"주인공은 Recall"

실제 암인데 암이라고 예측한 것(TP)도 좋아야함

근데 암이 아닌데 암이 아니라고 예측한 것(TN)도 좋아야함

---

TN = 특이도 (specity)

FP = 실제가 N 인데 틀린 것

FPR = 실제 중 자기자신(실제가 N)

TPR = 실제 중 자기자신(실제가 P)
 
---

(민감도, TPR, Recall) 과 특이도(TN)가 둘 다 높은 것이 좋은 것이다.

그러면 특이도 대신 FPR을 사용해도 되는데

그 때는 TPR이 높을 때 FPR이 낮아야 좋은 것이다.

---

X축이 커진다 = FPR의 비율이 높아진다. 

- 이미 학습이 끝난 모델인데 TPR과 FPR은 고정되있어야하는데

- 바꿔가면서 추이를 보겠다.
