불러오는데 실패

- 요청이 거절 당했을때 

    - why? 

        1.

        - 보내는 사람: 클라이언트 정보 (requests header)

        - 받는사람 정보 : 서버 정보(response header)

        2.

        - 브라우저는 오키 : 브라우저의 requests header 오키

        - 크롤러는 403 : 크롤러의 requests header 거절

        3.

        - 브라우저의 requests header 를 복붙해서 get하자.

        4.
        
        - 브라우저 requests header 정보중에서 

        1순위) user-agent, referrer

        2순위) cookies

        3순위) auth, authorization 등 권한을 의미하는 단어가 포함된 값

        4순위) 전부 다
